## 7. Scrape the details of Data science recruiters from naukri.com.

```python
import pandas as pd
from bs4 import BeautifulSoup
import requests
import selenium
from selenium import webdriver
import time
```

```python
driver=webdriver.Chrome("chromedriver.exe")
```

```python
driver.get("https://www.naukri.com/")
```

```python
search_btn1=driver.find_element_by_xpath("/html/body/div[1]/div[1]/div/ul[1]/li[2]/a/div")
search_btn1.click()
```

```python
url="https://www.naukri.com/hr-recruiters-consultants"
```

```python
driver.get(url)
```

```python
search_job=driver.find_element_by_xpath("//input[@class='sugInp']")
search_job.send_keys("Data Science")
```

```python
search_btn2=driver.find_element_by_xpath("//*[@id='qsbFormBtn']")
search_btn2.click()
```

```python
name=[]
des=[]
company1=[]
skills=[]
loc=[]
```

```python
titles_tag=driver.find_elements_by_xpath("//span[@class='fl ellipsis']")
for i in titles_tag:
    name.append(i.text)
name[0:5]
```

```python
titles_tag1=driver.find_elements_by_xpath("//span[@class='ellipsis clr']")
for i in titles_tag1:
    des.append(i.text)
des[0:5]
```

```python
titles_tag2=driver.find_elements_by_xpath("//*[@id='tabP-1']/div/div/div[1]/div[1]/p/a[2]")
for i in titles_tag2:
    company1.append(i.text)
company1[0:5]
```

```python
titles_tag3=driver.find_elements_by_xpath("//div[@class='hireSec highlightable']")
for i in titles_tag3:
    skills.append(i.text)
skills[0:5]
```

```python
titles_tag4=driver.find_elements_by_xpath("//small[@class='ellipsis']")
for i in titles_tag4:
    loc.append(i.text)
loc[0:5]
```

```python
import pandas as pd
Job=pd.DataFrame()
Job['Name']=name[0:49]
Job['Designation']=des[0:49]
Job['Company']=company1[0:49]
Job['Skills_they_hire_for']=skills[0:49]
Job['Location']=loc[0:49]
```
