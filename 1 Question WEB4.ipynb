## 1.Scrape the details of most viewed videos on YouTube from Wikipedia:

```python
# Importing Libraries
import pandas as pd
from bs4 import BeautifulSoup
import requests
import selenium
from selenium import webdriver
import time
```

```python
driver=webdriver.Chrome("chromedriver.exe")
```

```python
driver.get("https://en.wikipedia.org/wiki/List_of_most-viewed_YouTube_videos")
```

```python
rank=[]
name=[]
artist=[]
upload_date=[]
views=[]
```

```python
titles_tag=driver.find_elements_by_xpath("//*[@id='mw-content-text']/div[1]/table[3]/tbody/tr/td[1]")
for i in titles_tag:
    rank.append(i.text)
rank[0:9]
```

```python
titles_tag1=driver.find_elements_by_xpath("//*[@id='mw-content-text']/div[1]/table[3]/tbody/tr/td[2]/a")
for i in titles_tag1:
    name.append(i.text)
name[0:9]
```

```python
titles_tag3=driver.find_elements_by_xpath("//*[@id='mw-content-text']/div[1]/table[3]/tbody/tr/td[3]/a")
for i in titles_tag3:
    artist.append(i.text)
artist[0:9]
```

```python
titles_tag4=driver.find_elements_by_xpath("//*[@id='mw-content-text']/div[1]/table[3]/tbody/tr/td[5]")
for i in titles_tag4:
    upload_date.append(i.text)
upload_date[0:9]
```

```python
titles_tag5=driver.find_elements_by_xpath("//*[@id='mw-content-text']/div[1]/table[3]/tbody/tr/td[4]")
for i in titles_tag5:
    views.append(i.text)
views[0:9]
```

```python
import pandas as pd
YouTube=pd.DataFrame()
YouTube['Rank']=rank[0:29]
YouTube['Name']=name[0:29]
YouTube['Artist']=artist[0:29]
YouTube['Upload_Date']=upload_date[0:29]
YouTube['Views_Billions']=views[0:29]
```
